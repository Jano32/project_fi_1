```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_row', 111)
pd.set_option('display.max_column', 111)
```

```python
df = pd.read_excel("C:/Users/Jean-AntoineGRAU/Desktop/DataScientest/Projet/data_macro.xls")
df.head()
```

```python
df.info()
```

```python
df = df.dropna(subset = ['Prév.'])
df.info()
```

```python
df.rename(columns = {'Heure' : 'Hour', 'Devi.' : 'Forex', 'Événement' : 'Event', 'Prév.' : 'Forecast', 'Précédent' : 'Previous'},
          inplace = True)
df.info()
```

```python
df['Date'] = df['Date'].astype('string')
```

```python
df.info()
```

```python
df[['day', 'day-num', 'month', 'year']] = df['Date'].str.split(' ', expand = True)
df.head()
```

```python
df = df.drop(['Date', 'day'], axis = 1)
df.head()
```

```python
df['month'] = df['month'].replace(['janvier', 'février', 'mars', 'avril', 'mai', 'juin',
                                'juillet', 'août', 'septembre', 'octobre', 'novembre', 'décembre'],
                               ['01', '02', '03', '04', '05', '06',
                               '07', '08', '09', '10', '11', '12'])
df.head()
```

```python
df['Date'] = df[['day-num', 'month', 'year']].apply(' '.join, axis = 1)
df.head()
```

```python
df.drop(['day-num', 'month', 'year'], axis = 1, inplace = True)
df.head()
```

```python
first_column = df.pop('Date')
df.insert(0, 'Date', first_column)
df.head()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df.info()
```

```python
def is_percentage(x):
    if '%' in x:
        return True
    else:
        return False
   
    
df['is_percentage'] = df['Actuel'].apply(is_percentage)
df.head()
```
