1/ Suppresion des NA en fonction de la colonne 'Prév' qui sera la référence pour savoir si une publication est supérieure / inférieure / en ligne par rapport aux attentes

```python
df = df.dropna(subset = ['Prév.'])
df.info()
```

2/ Modification des noms des variables

```python
df.rename(columns = {'Heure' : 'Hour', 'Devi.' : 'Forex', 'Événement' : 'Event', 'Prév.' : 'Forecast', 'Précédent' : 'Previous'},
          inplace = True)
df.info()
```

3/ Travail de pre-processing

    3a.Changement du format de la colonne 'Date' en Datetime 

```python
df['Date'] = df['Date'].astype('string')
```

```python
df.info()
```

```python
df[['day', 'day-num', 'month', 'year']] = df['Date'].str.split(' ', expand = True)
df.head()
```

```python
df = df.drop(['Date', 'day'], axis = 1)
df.head()
```

```python
df['month'] = df['month'].replace(['janvier', 'février', 'mars', 'avril', 'mai', 'juin',
                                'juillet', 'août', 'septembre', 'octobre', 'novembre', 'décembre'],
                               ['01', '02', '03', '04', '05', '06',
                               '07', '08', '09', '10', '11', '12'])
df.head()
```

```python
df['Date'] = df[['day-num', 'month', 'year']].apply(' '.join, axis = 1)
df.head()
```

```python
df.drop(['day-num', 'month', 'year'], axis = 1, inplace = True)
df.head()
```

```python
first_column = df.pop('Date')
df.insert(0, 'Date', first_column)
df.head()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df.info()
```

```python
df.head()
```

3b. Nettoyage de la colonne 'Actuel' et modification du format en 'float'

```python
result = [i for i in df['Actuel'].dropna().unique()]
result = ''.join(result)
set(result)
```

```python
df[df['Actuel'].str.contains('%')]

#créer une colonne avec le type de valeur (est ce que c'est un %, un taux, une unité monétaire, etc)
```

```python
def is_percentage(x):
    if '%' in x:
        return True
    else:
        return False
   
    
df['is_percentage'] = df['Actuel'].apply(is_percentage)
df.head()
```

```python
df[df['is_percentage'] == True]
```

```python
def is_M(x):
    if 'M' in x:
        return True
    else:
        return False
   
    
df['is_M'] = df['Actuel'].apply(is_M)
df.head()
```

```python
def is_K(x):
    if 'K' in x:
        return True
    else:
        return False
   
    
df['is_K'] = df['Actuel'].apply(is_K)
df.head()
```

```python
def is_dot(x):
    if '.' in x:
        return True
    else:
        return False
   
    
df['is_dot'] = df['Actuel'].apply(is_dot)
df.head()
```

```python
df['Actuel'] = df['Actuel'].str.replace('.', '')
df[df['is_dot'] == True]
```

```python
df['Actuel'] = df['Actuel'].str.replace('K', '')
df[df['is_K'] == True]
```

```python
df['Actuel'] = df['Actuel'].str.replace('M', '')
df[df['is_M'] == True]
```

```python
df['Actuel'] = df['Actuel'].str.replace('%', '')
df[df['is_percentage'] == True]
```

```python
df['Actuel'] = df['Actuel'].str.replace(',', '.')
```

```python
df[df['is_percentage'] == True]
```

```python
df['Actuel'] = df['Actuel'].astype(float)
```

```python
df.loc[df.is_percentage == True, 'Actuel'] = df['Actuel'] / 100

```
